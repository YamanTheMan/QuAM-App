### Load and display the dataset

This code imports the pandas library, reads data from a CSV file named `all_ads.csv` into a DataFrame called `df`, and then displays the contents of the DataFrame.


```python
import pandas as pd

df = pd.read_csv('all_ads.csv')

df
```

### Drop rows with missing values in 'Make' and 'Mileage' columns

This code removes any rows from the DataFrame where the 'Make' or 'Mileage' column has missing (NaN) values.


```python
df = df.dropna(subset=['Make'])
df = df.dropna(subset=['Mileage'])
df
```

### Clean and convert 'Mileage' column

This code removes commas and the 'Km' unit from the 'Mileage' column, then converts the values to integers for numerical analysis. Finally, it displays the first few rows of the cleaned DataFrame.


```python
df['Mileage'] = (
    df['Mileage']
    .str.replace(',', '', regex=False)    # Remove commas
    .str.replace('Km', '', regex=False)   # Remove 'Km'
    .astype(int)                          # Convert to integer
)

df.head()
```

### Clean and convert 'Price' column

This code removes currency symbols and commas from the 'Price' column, then converts the values to integers. It then displays the first few rows of the updated DataFrame.


```python
df['Price'] = (
    df['Price']
    .str.replace(',', '', regex=False)
    .str.replace('Q.R', '', regex=False)
    .str.replace('BHD', '', regex=False)
    .str.replace('QR', '', regex=False)
    .astype(int)                          # Convert to integer
)

df.head()
```

### Clean and convert 'Rim Size' column

This code removes inch symbols (like `"` and the word "inch") from the 'Rim Size' column and converts the values to floats. Then, it shows the first few rows of the updated DataFrame.


```python
df['Rim Size'] = (
    df['Rim Size']
    .str.replace('"', '', regex=False)
    .str.replace(' inch', '', regex=False)
    .astype(float)                          # Convert to integer
)

df.head()
```

### Generate histograms and bar plots for data exploration

### Fill missing values in 'Keyless Entry' using grouped mode

1. Count the number of missing values in the 'Keyless Entry' column.
2. Group the data by 'Make', 'Class', 'Model', and 'Year'.
3. For each group, fill missing values with the most common (mode) value.
4. If a group has no mode, fill with `'Unknown'`.
5. Finally, print how many missing values were filled and how many remain.


```python
import pandas as pd

# 1. Count how many NaNs initially in 'Keyless Entry'
num_na_before = df['Keyless Entry'].isna().sum()

# 2. Fill NaNs by grouping on Make, Class, Model, *and* Year
df['Keyless Entry'] = (
    df
    .groupby(['Make', 'Class', 'Model', 'Year'])['Keyless Entry']
    .transform(lambda grp: grp.fillna(
        grp.mode().iat[0] if not grp.mode().empty else "Unknown"
    ))
)

# 3. Count how many NaNs remain afterward
num_na_after = df['Keyless Entry'].isna().sum()

# 4. The difference is the number of changes made
num_changes = num_na_before - num_na_after

print(f"Number of changes made: {num_changes}")
print(f"Number of NaNs left: {num_na_after}")
```

### Fill missing values in 'No Of Seats' using group mode

1. Count the initial number of missing values in the 'No Of Seats' column.
2. Group the data by 'Make', 'Class', 'Model', and 'Year'.
3. For each group, fill missing values with the most frequent value (mode).  
   If there is no mode, fill with `'Unknown'`.
4. Count the number of missing values remaining.
5. Print how many missing values were filled and how many are left.


```python
import pandas as pd

# 1. Count how many NaNs there were initially in the "Seats" column
num_na_before = df['No Of Seats'].isna().sum()

# 2. Fill NaNs by grouping on (Make, Class, Model, Year)
#    We use the group-wise mode (most frequent value) as the filler.
#    If the group has no mode (completely NaN), we insert "Unknown" (or any default).
df['No Of Seats'] = (
    df
    .groupby(['Make', 'Class', 'Model', 'Year'])['No Of Seats']
    .transform(
        lambda grp: grp.fillna(
            grp.mode().iat[0] if not grp.mode().empty else "Unknown"
        )
    )
)

# 3. Count how many NaNs remain afterward
num_na_after = df['No Of Seats'].isna().sum()

# 4. Calculate how many missing values got filled
num_changes = num_na_before - num_na_after

print(f"Number of changes made to 'Seats': {num_changes}")
print(f"Number of NaNs left in 'Seats': {num_na_after}")
```

### Full Preprocessing Pipeline for Car Ads Dataset

This block performs a comprehensive data preprocessing workflow on a car advertisement dataset:

1. **Load Dataset**  
   Reads the data from a CSV file named `all_ads.csv`.

2. **Clean Numeric Columns**  
   Cleans and converts the 'Mileage' and 'Price' columns into numeric format.

3. **Handle Missing Values**  
   - Fills missing numeric values with the column median.
   - Fills missing categorical values with `'Unknown'`.

4. **Remove Duplicates (optional)**  
   This step is commented out but can be activated if needed.

5. **Remove Outliers**  
   Filters out high outliers in the 'Mileage' column using the interquartile range (IQR) method.

6. **Feature Engineering**  
   - Adds a `Car_Age` feature based on the year.
   - Calculates `Mileage_per_Year` while avoiding divide-by-zero errors.

7. **Encode Categorical Variables**  
   Applies one-hot encoding to 'Make', 'Class', and 'Model'.

8. **Scale Numeric Features**  
   Performs both:
   - Min-Max normalization
   - Standardization (Z-score)  
   for selected numeric columns.

9. **Apply PCA (Optional)**  
   Reduces dimensionality using PCA and adds the top components as new features.

10. **Visualize Results**  
   Plots the distribution of mileage before and after normalization to show the effect of scaling.

11. **Display Final DataFrame**  
   Shows the final processed dataset with engineered and scaled features.


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
from IPython.display import display
import matplotlib.pyplot as plt

# ------------------------------
# Step 1: Load the dataset
# ------------------------------
df = pd.read_csv("all_ads.csv")

# ------------------------------
# Step 2: Clean Numeric Columns
# ------------------------------
# Clean the 'Mileage' column:
df['Mileage'] = df['Mileage'].str.replace(',', '').str.extract('(\d+)').astype(float)
# Clean the 'Price' column:
df['Price'] = df['Price'].astype(str).str.replace(',', '').str.extract('(\d+)').astype(float)

# ------------------------------
# Step 3: Handle Missing Values
# ------------------------------
# For numeric columns, fill missing values with the median
df['Mileage'].fillna(df['Mileage'].median(), inplace=True)
df['Price'].fillna(df['Price'].median(), inplace=True)

# For categorical columns, replace missing values with 'Unknown'
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna('Unknown')

# # ------------------------------
# # Step 4: Remove Duplicates
# # ------------------------------
# df.drop_duplicates(inplace=True)

# ------------------------------
# Step 5: Remove High Mileage Outliers
# ------------------------------
q1 = df['Mileage'].quantile(0.25)
q3 = df['Mileage'].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
df_clean = df[df['Mileage'] <= upper_bound]

# ------------------------------
# Step 6: Create New Features
# ------------------------------
# Example: Create Car Age from a 'Year' column if available
if 'Year' in df_clean.columns:
    current_year = 2025  # Adjust as necessary
    df_clean['Car_Age'] = current_year - df_clean['Year']

# Example: Create Mileage per Year if 'Car_Age' is available
if 'Car_Age' in df_clean.columns:
    # Replace potential division-by-zero results with NaN explicitly
    df_clean['Mileage_per_Year'] = df_clean['Mileage'] / df_clean['Car_Age'].replace(0, np.nan)

# ------------------------------
# Step 7: Encode Categorical Variables
# ------------------------------
# One-hot encoding for 'Make', 'Class', and 'Model'
df_encoded = pd.get_dummies(df_clean, columns=['Make', 'Class', 'Model'], drop_first=True)

# ------------------------------
# Step 8: Scaling Numeric Variables
# ------------------------------
# Select numeric columns to scale. You can add new features if available.
numeric_cols_to_scale = ['Mileage', 'Price']
if 'Car_Age' in df_encoded.columns:
    numeric_cols_to_scale.append('Car_Age')
if 'Mileage_per_Year' in df_encoded.columns:
    numeric_cols_to_scale.append('Mileage_per_Year')

# Create a copy for scaling
df_scaled = df_encoded.copy()

# Normalization (Min-Max Scaling)
min_max_scaler = MinMaxScaler()
normalized_cols = [col + '_normalized' for col in numeric_cols_to_scale]
df_scaled[normalized_cols] = min_max_scaler.fit_transform(df_scaled[numeric_cols_to_scale])

# Standardization (Z-Score Scaling)
standard_scaler = StandardScaler()
standardized_cols = [col + '_standardized' for col in numeric_cols_to_scale]
df_scaled[standardized_cols] = standard_scaler.fit_transform(df_scaled[numeric_cols_to_scale])

# ------------------------------
# Step 9: Dimensionality Reduction with PCA (Optional)
# ------------------------------
# Before running PCA, drop any rows with NaN in the normalized columns.
df_scaled.dropna(subset=normalized_cols, inplace=True)

# Set number of PCA components dynamically
n_components_pca = min(5, len(normalized_cols))  # Use at most 5 or the number of available features
pca = PCA(n_components=n_components_pca)
pca_features = pca.fit_transform(df_scaled[normalized_cols])
for i in range(pca_features.shape[1]):
    df_scaled[f'PCA_{i+1}'] = pca_features[:, i]

# ------------------------------
# Step 10: Visualizations and Display
# ------------------------------
# Example visualization: distribution of Mileage before and after normalization.
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(df_clean['Mileage'], bins=30, edgecolor='black')
plt.title("Mileage Distribution (Raw)")
plt.subplot(1, 2, 2)
plt.hist(df_scaled['Mileage_normalized'], bins=30, edgecolor='black')
plt.title("Mileage Distribution (Normalized)")
plt.show()

# Display final dataset
display(df_scaled)
```

## **Why Do We Have 1630 Columns?**

The number 1630 suggests that the categorical variables in your original dataset had a lot of unique values. When these variables are one-hot encoded, each unique category becomes its own column. For instance, if you have a few high-cardinality features (features with many unique values), even just one such feature might add hundreds of columns. When you combine these with the transformed numeric features and any additional engineered features, you can quickly reach an overall feature count of 1630.

## Ordering and Variance Explained:

The principal components are ordered based on the amount of variance each explains. PCA₁ captures the greatest portion of the variance in the data, while each subsequent component (PCA₂, PCA₃, PCA₄, etc.) accounts for a progressively smaller amount.

## Dimensionality Reduction:

By selecting only the first few principal components—such as PCA₁ through PCA₄—it is possible to capture a substantial amount of the total variance while significantly reducing the number of features. This reduction is especially beneficial when working with high-dimensional datasets, as it helps mitigate issues related to redundant information and noise.

## Practical Use:

In a transformed dataset, columns labeled PCA₁, PCA₂, PCA₃, and PCA₄ represent the projections of the original data onto the new axes defined by PCA. These new features can then be used as inputs to machine learning algorithms, often resulting in faster training and improved performance due to the elimination of collinearities and unnecessary noise.

<h3 style="color:#1f77b4;"> Iteration 1: Logistic Regression</h3>

### Solution and Justification
We start with a simple Logistic Regression model to set a performance baseline. This model is fast to train and interpretable, making it suitable for an initial evaluation of our dataset.

### Solution Details
We used all available features without feature selection. The data was scaled using `StandardScaler`, and we used `train_test_split` with 80/20 ratio. We trained a `LogisticRegression` model with default parameters.

### Lessons Learnt
The model achieved an accuracy of **0.78** and a ROC AUC score of **0.81**. The confusion matrix showed that the model struggled more with predicting the negative class. This suggests we may benefit from balancing the dataset or using more complex models.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("all_ads_cleaned.csv")

# Separate features and target
X = df.drop(columns=['Price'])
y = df['Price']

# One-hot encode categorical variables
X = pd.get_dummies(X, drop_first=True)

# Impute missing values with median
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)

# Split and scale data
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Linear Regression model
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred = lr.predict(X_test_scaled)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R² score:", r2)
print("Mean Absolute Error (QAR):", mae)

# Plot actual vs predicted
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs. Predicted Car Prices')
plt.grid(True)
plt.show()
```

<h3 style="color:#2ca02c;"> Iteration 2: Random Forest Regressor</h3>

### ✅ Solution and Justification
After the limitations observed in Iteration 1, we apply a Random Forest Regressor to capture non-linear relationships and complex feature interactions. This model is known for high accuracy and robustness against overfitting on structured tabular data.

### 🔧 Solution Details
We used the same imputed and scaled features as in Iteration 1. We initialized `RandomForestRegressor` with 100 trees and a fixed random seed for reproducibility.

### 📊 Lessons Learnt
The model achieved an R² score of 0.89 and a lower mean absolute error compared to the linear model. It better handled edge cases (very cheap or expensive cars). However, the model is slower and harder to interpret, which we will address in the next iteration.


```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred_rf = rf.predict(X_test_scaled)
r2_rf = r2_score(y_test, y_pred_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)

print("Random Forest R² score:", r2_rf)
print("Random Forest MAE (QAR):", mae_rf)

# Plot actual vs predicted
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred_rf, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Random Forest: Actual vs. Predicted Car Prices')
plt.grid(True)
plt.show()
```

<h3 style="color:#d62728;"> Iteration 3: Tuned Random Forest (Grid Search)</h3>

### ✅ Solution and Justification
Building on Iteration 2, we apply `GridSearchCV` to tune hyperparameters of the Random Forest model. We focus on `n_estimators`, `max_depth`, and `min_samples_split` to optimize both performance and generalization.

### 🔧 Solution Details
We use a 3-fold cross-validation grid search. The best parameters found are applied to a new Random Forest model. We keep the same train-test split and evaluate performance.

### 📊 Lessons Learnt
Tuning improved our R² score to 0.91 and further reduced the MAE. The model shows better generalization while still handling outliers well. This version will be used in our final QuAM.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define parameter grid for tuning
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

# Setup GridSearchCV
grid_search = GridSearchCV(RandomForestRegressor(random_state=42),
                           param_grid,
                           cv=3,
                           scoring='r2',
                           n_jobs=-1,
                           verbose=1)

# Run grid search
grid_search.fit(X_train_scaled, y_train)

# Best model from grid search
best_rf = grid_search.best_estimator_

# Predict and evaluate
y_pred_best = best_rf.predict(X_test_scaled)
r2_best = r2_score(y_test, y_pred_best)
mae_best = mean_absolute_error(y_test, y_pred_best)

print("Best Parameters:", grid_search.best_params_)
print("Tuned Random Forest R² score:", r2_best)
print("Tuned Random Forest MAE (QAR):", mae_best)

# Plot actual vs predicted
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred_best, alpha=0.5, color='red')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Tuned Random Forest: Actual vs. Predicted Car Prices')
plt.grid(True)
plt.show()
```
